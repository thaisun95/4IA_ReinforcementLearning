```python
import sys
import os
sys.path.append(os.path.abspath(".."))

import numpy as np
import matplotlib.pyplot as plt
from rl_algorithms.temporal_forces.sarsa import sarsa, get_policy_from_q, evaluate_policy
```

```python
from envs.line_world.line_world import LineWorld
import matplotlib.pyplot as plt

print("=== SARSA sur Line World ===")
env = LineWorld(size=5, start_state=2)
Q, rewards, lengths = sarsa(env, episodes=500, verbose=False)
plt.plot(rewards)
plt.title("Line World - Récompense par épisode")
plt.show()
policy = get_policy_from_q(Q, env)
print("Politique optimale:", policy)
avg_reward, avg_length = evaluate_policy(env, policy)
print(f"Reward moyen: {avg_reward:.2f}, Longueur moyenne: {avg_length:.2f}\n")
```

```python
from envs.grid_world.grid_world import GridWorld

print("=== SARSA sur Grid World ===")
env = GridWorld(n_rows=2, n_cols=2, start_state=(0,0))
Q, rewards, lengths = sarsa(env, episodes=300, alpha=0.2, gamma=0.99, epsilon=0.1)
plt.plot(rewards)
plt.title("Grid World - Récompense par épisode")
plt.show()
policy = get_policy_from_q(Q, env)
env.render_policy(policy)
avg_reward, avg_length = evaluate_policy(env, policy)
print(f"Reward moyen: {avg_reward:.2f}, Longueur moyenne: {avg_length:.2f}\n")
```
